```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def network(N):
    P = np.zeros((3, N+1))
    P[0][0] = 1
    for i in range(1, N+1):
        P[0][i] = 0.5*P[1][i-1] + P[2][i-1]
        P[1][i] = P[0][i-1]
        P[2][i] = 0.5*P[1][i-1]
    return P
```

```python
n = 100
P = network(n)

plt.figure()
plt.plot([x for x in range(n+1)], P[0], label = r"$P(X_n=A)$")
plt.plot([x for x in range(n+1)], P[1], label = r"$P(X_n=B)$")
plt.plot([x for x in range(n+1)], P[2], label = r"$P(X_n=C)$")
plt.xlabel("n")
plt.ylabel("probability")
plt.legend()
plt.show()
```
