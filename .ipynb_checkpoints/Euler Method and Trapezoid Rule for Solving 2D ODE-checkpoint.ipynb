# Euler Method and Trapezoid Rule for Solving 2D ODE

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.linalg
```

```python
# u0: np.array([a, b])
# A: np.array([[a11, a12],[a21, a22]])
def Euler(u0, A, dt, N):
    I = np.array([[1, 0],[0, 1]])
    u = np.zeros((2, N+1))
    u[:,0] = u0
    for i in range(1, N+1):
        u[:,i] = (I + dt*A)@u[:,i-1]
    return u
```

```python
def Trapezoid(u0, A, dt, N):
    I = np.array([[1, 0],[0, 1]])
    M = np.linalg.inv(I-dt/2*A)@(I+dt/2*A)
    u = np.zeros((2, N+1))
    u[:,0] = u0
    for i in range(1, N+1):
        u[:,i] = M@u[:,i-1]
    return u
```

```python
# Initialization
u0 = np.array([3, 3])
A = np.array([[0, -2],[2, 0]])
dt = 0.01
N = 1000

# Find Numerical Solution
u1 = Euler(u0, A, dt, N)
u2 = Trapezoid(u0, A, dt, N)

# Calculate Exact Solution
u_exact = np.zeros((2, N+1))
for i in range(N+1):
    u_exact[:,i] = scipy.linalg.expm(dt*i*A)@u0
    

plt.figure()
plt.plot(u1[0,:],u1[1,:],'g',label='Euler')
plt.plot(u2[0,:],u2[1,:],'b',label='Trapezoid')
plt.plot(u_exact[0,:],u_exact[1,:],'r:',label='Exact Solution')
plt.legend()
plt.show()
```

## Manipulating with Errors

Let's first have a look at the errors from the two numerical methods with our initial choice of $\Delta t = 0.01$.\
Note that we're using the Euclidean norm to calculate the errors.

```python
T = [dt*i for i in range(N+1)]
error_E = list()
error_T = list()
for j in range(len(T)):
    x_exact = u_exact[:,j]
    x_E = u1[:,j]
    x_T = u2[:,j]
    error_E.append(np.sum(np.square(x_exact - x_E)))
    error_T.append(np.sum(np.square(x_exact - x_T)))
    
plt.figure()
plt.plot(T, error_E, label='Euler')
plt.plot(T, error_T, label='Trapezoid')
plt.title('Errors of Numerical Solution with dt=0.01')
plt.xlabel('t')
plt.ylabel('error')
plt.legend()
plt.show()
```

Let's make $\Delta t$ smaller to obtain more accurate solutions, we now set $\Delta t = 0.001$.

```python
dt2 = 0.001
N2 = int(10/dt2)
uE = Euler(u0, A, dt2, N2)
uT = Trapezoid(u0, A, dt2, N2)
u_exact2 = np.zeros((2, N2+1))
for i in range(N2+1):
    u_exact2[:,i] = scipy.linalg.expm(dt2*i*A)@u0

plt.figure()
plt.plot(uE[0,:],uE[1,:],'g',label='Euler')
plt.plot(uT[0,:],uT[1,:],'b',label='Trapezoid')
plt.plot(u_exact[0,:],u_exact[1,:],'r:',label='Exact Solution')
plt.legend()
plt.show()
```

```python
T2 = [dt2*i for i in range(N2+1)]
error_E2 = list()
error_T2 = list()
for j in range(len(T2)):
    x_exact = u_exact2[:,j]
    x_E = uE[:,j]
    x_T = uT[:,j]
    error_E2.append(np.sum(np.square(x_exact - x_E)))
    error_T2.append(np.sum(np.square(x_exact - x_T)))
    
plt.figure()
plt.plot(T2, error_E2, label='Euler')
plt.plot(T2, error_T2, label='Trapezoid')
plt.title('Errors of Numerical Solution with dt=0.001')
plt.xlabel('t')
plt.ylabel('error')
plt.legend()
plt.show()
```
