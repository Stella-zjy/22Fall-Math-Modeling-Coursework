# Fixed Point Iteration

```python
import numpy as np
import matplotlib.pyplot as plt
```

To find the roots of $sinx - 0.5x = 0$ is the same as finding the fixed points of $f(x) = 2sin(x)$.

```python
x = np.linspace(-3,3,1000)
y = [2*np.sin(i) for i in x]
plt.figure()
plt.plot(x, x, label='y = x')
plt.plot(x, y, label='y = 2sin(x)')
plt.legend()
plt.grid()
plt.show()
```

The above graph gives a general idea about where the target points are, to find those nonzero ones, we may then perform fixed point iteration starting with $x_0 = 2$ and $x_0 = -2$ respectively.

```python
def fixed_point_iter(f, x0, e):
    x = x0
    i = 0
    while abs(f(x)-x) >= e:
        x = f(x)
        i += 1
    return x, i
```

In our case, in order to find points that are accurate to 9 decimals, we will need 48 iterations in total.

```python
f = lambda x: 2*np.sin(x)
x1,i1 = fixed_point_iter(f, 2, 1e-10)
x2,i2 = fixed_point_iter(f, -2, 1e-10)
print("The first nonzero point we get is", round(x1,9),"after iteration", i1)
print("The second nonzero point we get is", round(x2,9),"after iteration", i2)
```
